```python
import pandas as pd
import matplotlib as mlp
from matplotlib import pyplot as plt
import seaborn as sns
import missingno as msno
import numpy as np
```

```python
#pokold = pd.read_csv("/mnt/c/Users/Georg/Desktop/aiproject2022/Pokemon.csv")

pokold = pd.read_csv("/mnt/c/Users/Kevin/PokemonProject/Pokemon.csv")
```

```python
pok = pokold.rename(columns={'Sp. Def':'SpecialDefense', 'Sp. Atk' : 'SpecialAttack',"Type 1": "Type1", "Type 2": "Type2"})
pok
```

```python
pok.insert(11,'Attack1',value = 0, allow_duplicates=True)
pok.insert(12,"Defense1",value = 0, allow_duplicates = True)
pok.insert(13,'SpecialAttack1',value = 0, allow_duplicates = True)
pok.insert(14,'SpecialDefense1',value = 0, allow_duplicates = True)
pok.insert(15,"Speed1",value = 0, allow_duplicates = True)
pok.insert(6,"HP1",value = 0, allow_duplicates = True)


pok.HP1 = pok.HP/pok.Total
pok.Attack1 = pok.Attack/pok.Total
pok.Defense1 = pok.Defense/pok.Total
pok.SpecialAttack1 = pok.SpecialAttack/pok.Total
pok.SpecialDefense1 = pok.SpecialDefense/pok.Total
pok.Speed1 = pok.Speed/pok.Total

del pok["HP"]
del pok["Attack"]
del pok["Defense"]
del pok ["SpecialAttack"]
del pok["SpecialDefense"]
del pok["Speed"]
pok
```

```python
pok = pok.round({"HP1": 4, "Attack1": 4, "Defense1": 4, "SpecialAttack1":4, "SpecialDefense1": 4, "Speed1": 4})
pok
```

```python
pokcount = pok["Type1"].value_counts() + pok["Type2"].value_counts()



alltype = pd.concat([pok["Type1"], pok["Type2"]])
```

```python
fig, ax = plt.subplots(figsize = (14,8))
sns.countplot(x = alltype, ax = ax, palette = "viridis", order = alltype.value_counts().index )

```

```python
from matplotlib import cm
pokavg = pok[["Type1", "HP1", "Attack1", "Defense1", "SpecialAttack1", "SpecialDefense1", "Speed1"]]

stats = ['HP1', 'Attack1', 'Defense1', 'SpecialAttack1', 'SpecialDefense1', 'Speed1']


for x in stats: 
    pok_avg = pokavg.groupby(["Type1"])[x].mean().reset_index()

    pok_avg = pok_avg.sort_values([x], ascending = True).plot(kind = 'bar', x = 'Type1', title = "Type 1 Element vs " + x)
    
    print(pok_avg)



```

```python
pokavg2  = pok[["Type2", "HP1", "Attack1", "Defense1", "SpecialAttack1", "SpecialDefense1", "Speed1"]]


for x in stats:
    
    pok2_avg = pokavg2.groupby(["Type2"])[x].mean().reset_index()

    pok2_avg = pok2_avg.sort_values([x], ascending = True).plot(kind = 'bar', x = 'Type2', title = "Type 2 Element vs " + x)
    
    print(pok2_avg)
```

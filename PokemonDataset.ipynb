```python
import pandas as pd
import matplotlib as mlp
from matplotlib import pyplot as plt
import seaborn as sns
import missingno as msno
import numpy as np
```

```python
pokold = pd.read_csv("/mnt/c/Users/Georg/Desktop/aiproject2022/Pokemon.csv")
```

```python
pok = pokold.rename(columns={'Sp. Def':'SpecialDefense', 'Sp. Atk' : 'SpecialAttack',"Type 1": "Type1", "Type 2": "Type2"})
pok
```

```python
pok.insert(11,'Attack1',value = 0, allow_duplicates=True)
pok.insert(12,"Defense1",value = 0, allow_duplicates = True)
pok.insert(13,'SpecialAttack1',value = 0, allow_duplicates = True)
pok.insert(14,'SpecialDefense1',value = 0, allow_duplicates = True)
pok.insert(15,"Speed1",value = 0, allow_duplicates = True)
pok.insert(6,"HP1",value = 0, allow_duplicates = True)


pok.HP1 = pok.HP/pok.Total
pok.Attack1 = pok.Attack/pok.Total
pok.Defense1 = pok.Defense/pok.Total
pok.SpecialAttack1 = pok.SpecialAttack/pok.Total
pok.SpecialDefense1 = pok.SpecialDefense/pok.Total
pok.Speed1 = pok.Speed/pok.Total

del pok["HP"]
del pok["Attack"]
del pok["Defense"]
del pok ["SpecialAttack"]
del pok["SpecialDefense"]
del pok["Speed"]
pok
```

```python
pok = pok.round({"HP1": 4, "Attack1": 4, "Defense1": 4, "SpecialAttack1":4, "SpecialDefense1": 4, "Speed1": 4})
pok
```
